```python
#Q No 1...
MyFriendData = {
    "first_name": "Usama",
    "last_name" : "Javaid",
    "age"  : 20,
    "city" : "Hyderabad"
}

update_dic = {"Qualification" : "Undergraduate"} 

print(MyFriendData)

# add new key and value in a disctionary  
MyFriendData.update(update_dic)

## After Update Print MyFriendData Dictionary
print("\nAfter Update :\n",  MyFriendData)

## Delete the Update Key and Value
del MyFriendData["Qualification"]



## After Delete 'Update Key and Value from MyFriendData' Dictionary
print("\nAfter Delete :\n",  MyFriendData)





#Q No 2...
Cities = {
    "City1" : "Islamabad",
    "City2" : "Lahore"   ,
    "City3" :"Hyderabad"  
}

Information_Of_Cities = {
    "City1"                 : "Capital of pakistan",  
    "populationofislamabad" : "1.5 Crore",
    "fact_islamabad"        : "Most Beautiful City of Pakistan",    
    
    "City2"                 : "Lahore",  
    "populationoflahore"    : "2 Crore",
    "fact_lahore"           : "Famous for it's Food",    

    "City3"                 : "Hyderabad",  
    "populationofhyderabad" : "1.8 Crore",
    "fact_hyderabad"        : "It is the second largest city of Sindh."    
    
}

print("\n\n\n\n",Cities)
print(" ")
print(Information_Of_Cities)





#Q No 3...
personsAge = int(input("\n\n\n\nEnter Person's Age: "))

if(personsAge <=3 ):
    print("The Ticket Is Free For Only You...")
    
elif(personsAge>3 and personsAge<=12):
        print("Ticket Is $10...")
        
elif(personsAge>12):
            print("Ticket Is $15...")


        
        
#Q No 4...
def favourite_book(title):

    print(title)

favourite_book("\n\n\n\nOne of my favorite books is 'Let Us C Programming' ")





#Q No 5...
### import Module of Random Number
from random import randint

### Assign Module = any data type
RandomNumber = randint(1,30)
#print(RandomNumber)

### 3 Chances for User Guess the Number... 
chances = 3
print("\n\n\n'Alert...' Total Chances = 3 ...! ")

### while loop for running game all time
### Until User guess the Correct Number...
keep_playing = "true"
while keep_playing == "true":
    
    guess = int(input("What Is Your Guess No Between '1 To 30 Numbers'... "))
    print(" ")

    ## Decreases Chances -1 Every time
    chances = chances -1
    
    ## Guess Equal to the Random Number so 'User' Won...
    if guess == RandomNumber:
        print("\n\nYou Correctly Guess The Number", "\nCongrats You've Won the Game...!\n")
        keep_playing = "false"

    else:    
        if guess > RandomNumber:     ## if guess is Greater than Random Number..
            print("\nYour Guess is Too High")
            print("Remaining Chances : ", chances)
            
        elif guess < RandomNumber:   ## if guess is Less than Random Number..
            print("\nYour Guess is Too Low")
            print("Remaining Chances : ", chances)
    
    ## Chances is Equal to Zero (0) So Game Over... 
    if chances == 0:
        print("You Have No More Chances Avilable")
        print("'Game Over' \t The Winning Number was: ", "\t", RandomNumber)
        keep_playing = "false"
        
```
