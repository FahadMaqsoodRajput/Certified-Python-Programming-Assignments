```python
#Q No 1...
print("\n\n\nDefine OOP ?")
print("\n-> Stands for 'Object-Oriented Programming.' ", 
      "\n-> OOP refers to a Programming Methodology based on 'Classes', 'Objects', instead of just 'Functions' ", 
      "and 'Procedures.'")
print("->These objects are organized into classes, which allow individual objects to be grouped together.",
      "\n-> Most modern programming languages including Java, C/C++, and PHP, are object-oriented languages" , 
      "and many older programming languages now have object-oriented versions.")
print("-> An 'object' in an OOP language refers to a specific type, or 'instance,'' of a class.")
print("-> Object-oriented programming makes it easier for programmers to structure and organize software programs.", 
      "Because individual objects can be modified without affecting other aspects of the program", 
      "it is also easier to update and change programs written in object-oriented languages. ")




#Q No 2...
print("\n\n\nList Down the Benefits of OOP :")
print("\nFollowing Are Few Benefits Of Object Oriented Programming: ")
print("\n-> Easier Troubleshooting", 
      "\n-> Flexibility through Polymorphism",  
      "\n-> Effective Problem Solving",
      "\n-> Re-Usability", 
      "\n-> Data Redundancy",
      "\n-> Code Maintenance",
      "\n-> Security",
      "\n-> Design Benefits",
      "\n-> Better Productivity",
      "\n-> Problems Solving")




#Q No 3
print("\n\n\nDifference Between 'Function' And 'Method' ?")
print("\nFunction:-")
print("\n\t-> A function is a block of code to carry out a specific task", 
             "will contain its own scope and is called by name.", 
      "\n\t-> All functions may contain zero(no) arguments or more than one arguments.")
print("\t-> Function is block of code that is also called by its name. (independent)")
print("\t-> The function can have different parameters or may not have any at all. ")
print("\t-> It may or may not return any data.",
      "\n\t-> Function does not deal with Class and its instance concept.")
print("\tBelow Defined Two Types Of Functions in Python:",
      "\n\te.g -> User-Defined Function, \n\t    -> Inbuilt Function.")


print("\nMethod:-")
print("\n\t-> A method in python is somewhat similar to a function, except it is associated with object/classes.", 
      "\n\t-> Methods in python are very similar to functions except for two major differences.")
print("\t-> Method is called by its name, but it is associated to an object (dependent).")
print("\t-> It may or may not return any data.")
print("\tBelow Defined Two Types Of Methods in Python: ",
      "\n\te.g -> User Defined Method,   \n\t    -> Inbuilt Method.")





#Q No 4
## (1) 
print("\n\n\nDefine Class ?")
print("\nA Class is a Piece of the Program's source code that describes a particular type of Objects.")
print("\t\tOR" , "\nA Class is a Template.. Whenever its Object is Created it holds No Memory Space. ")
print("\t\tOR" , "\nA Blueprint for Objects of a particular type." , 
                 "\nClass also Defines available behaviors of its Objects.")


## (2) 
print("\n\n\nDefine Object ? ")
print("\nAn Object is called 'An Instance of a Class' ")
print("When an Object is Created, a chunk of memory is allocated to hold it's 'Attributes' And 'Behaviors'.")
   

    
## (3) 
print("\n\n\nDefine Attribute ? ")
print("\nAttributes are frequently referred to as members or properties.")
print("\tOR", "\nA class attribute is a Python variable that belongs to a class rather than a particular object.", 
          "It is shared between all the objects of this class and", 
          "It is defined outside the constructor function, ' __init__(self,...) ' ,of the class.")
          

## (3) 
print("\n\n\nDefine Behavior ?")          
print("\nBehaviors are actions that can occur on an object.", 
      "\nThe behaviors that can be performed on a specific class of object are called methods. ")





## Q No 5...  
class Car():
    def __init__(self, model, color, name, gear, speed):
        self.model = model
        self.color = color
        self.name  = name 
        self.gear = gear
        self.speed = speed

##Car 1 Properties        
carProperties = Car(2014, "Black", "Corolla", 5, "200 Km/h")
print("\n\n\n\n\n\nCar 1 Instance: \n")
print("Model Year of Car Is: ", carProperties.model)
print("Color of Car Is: ", carProperties.color)
print("Name of Car Is: ", carProperties.name)
print("Gear in This Car : ", carProperties.gear)
print("Speed of This Car Is: ", carProperties.speed)

##Car 2 Properties
carProperties_2 = Car(2012, "White", "Suzuki Mehran", 5, "180 Km/h")
print("\n\nCar 2 Instance: \n")
print("Model Year of Car Is: ", carProperties_2.model)
print("Color of Car Is: ", carProperties_2.color)
print("Name of Car Is: ", carProperties_2.name)
print("Gear in This Car : ", carProperties_2.gear)
print("Speed of This Car Is: ", carProperties_2.speed)


##Car 3 Properties
carProperties_3 = Car(2002, "White", "", 5, "150 Km/h")
print("\n\nCar 3 Instance: \n")
print("Model Year of Car Is: ", carProperties_3.model)
print("Color of Car Is: ", carProperties_3.color)
print("Name of Car Is: ", carProperties_3.name)
print("Gear in This Car : ", carProperties_3.gear)
print("Speed of This Car Is: ", carProperties_3.speed)



##Car 4 Properties
carProperties_4 = Car(2017, "Blue", "City", 5, "220 Km/h")
print("\n\nCar 4 Instance: \n")
print("Model Year of Car Is: ", carProperties_4.model)
print("Color of Car Is: ", carProperties_4.color)
print("Name of Car Is: ", carProperties_4.name)
print("Gear in This Car : ", carProperties_4.gear)
print("Speed of This Car Is: ", carProperties_4.speed)


##Car 5 Properties
carProperties_5 = Car(2018, "Red", "Ferrari", 6, "350 Km/h")
print("\n\nCar 5 Instance: \n")
print("Model Year of Car Is: ", carProperties_5.model)
print("Color of Car Is: ", carProperties_5.color)
print("Name of Car Is: ", carProperties_5.name)
print("Gear in This Car : ", carProperties_5.gear)
print("Speed of This Car Is: ", carProperties_5.speed)



```
